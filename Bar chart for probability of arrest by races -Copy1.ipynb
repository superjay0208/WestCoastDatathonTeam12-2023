```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\Downloads\West Coast 2023 Datasets\Traffic, Investigations _ Other\investigations.csv")
```

```python
df.groupby('race').count()
```

```python
temp_df = df.loc[(df['individual_arrested'] != 0)]
temp_df.groupby('race').count()
```

```python
proportion_of_arrest = temp_df.groupby('race').count()["individual_arrested"]/df.groupby('race').count()["individual_arrested"]
proportion_of_arrest
```

```python
race = ['American Indian', 'Asian', 'Black-Latino', 'Black-Non-Latino', 'Unknown', 'White-Latino', 'White-Non-Latino']
fig, ax = plt.subplots(figsize=(10, 10))
ax.bar(race, proportion_of_arrest)
plt.show()
```
